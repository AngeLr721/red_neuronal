```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.optimizers import RMSprop, SGD
from tensorflow.keras import regularizers
```

```python
eta = 0.01
epochs = 30
batch_size = 10
```

```python
dataset=mnist.load_data()
```

```python
dat=np.array(dataset)
print(dat[1,1].shape)
(x_train, y_train), (x_test, y_test) = dataset
```

```python
x_trainv = x_train.reshape(60000, 784)
x_testv = x_test.reshape(10000, 784)
x_trainv = x_trainv.astype('float32')
x_testv = x_testv.astype('float32')

x_trainv/=225
x_testv/=225
```

```python
num_classes=10
y_trainc = keras.utils.to_categorical(y_train, num_classes)
y_testc = keras.utils.to_categorical(y_test, num_classes)
```

```python
model = Sequential()
model.add(Dense(30, activation='sigmoid', input_shape=(28*28,)))
model.add(Dense(30, activation='sigmoid'))
model.add(Dense(num_classes, activation='sigmoid'))
model.summary()
model.compile(loss='mean_squared_error',optimizer=SGD(learning_rate=eta),metrics=['accuracy'])
```

```python
history = model.fit(x_trainv, y_trainc,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_testv, y_testc)
                    )

# Obtener la precisin y la prdida del historial de entrenamiento
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Crear las grficas
plt.figure(figsize=(12, 4))

# Grfica de precisin
plt.subplot(1, 2, 1)
plt.plot(range(1, len(train_accuracy) + 1), train_accuracy, label='Precisin en Entrenamiento')
plt.plot(range(1, len(val_accuracy) + 1), val_accuracy, label='Precisin en Validacin')
plt.title('Precisin vs. pocas')
plt.xlabel('pocas')
plt.ylabel('Precisin')
plt.legend()

# Grfica de prdida
plt.subplot(1, 2, 2)
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Prdida en Entrenamiento')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Prdida en Validacin')
plt.title('Prdida vs. pocas')
plt.xlabel('pocas')
plt.ylabel('Prdida')
plt.legend()

# Mostrar las grficas
plt.tight_layout()
plt.show()
```
